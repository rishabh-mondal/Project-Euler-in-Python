```python
def fib(n):
    if(n==1 or n==2):
        return n
    else:
        return (fib(n-1)+fib(n-2))
    
```

```python
def fibonacci(n, memo={}):
    if n in memo:
        return memo[n]
    if n == 1:
        return 1
    if n == 2:
        return 2
    memo[n] = fibonacci(n - 1, memo) + fibonacci(n - 2, memo)
    return memo[n]

def sum_even_fibonacci(limit):
    sum_even = 0
    n = 1
    while True:
        fib = fibonacci(n)
        if fib % 2 == 0:
            sum_even += fib
        n += 1
    return sum_even

# Define the limit
limit = 4000000
result = sum_even_fibonacci(limit)
print(result)


```
